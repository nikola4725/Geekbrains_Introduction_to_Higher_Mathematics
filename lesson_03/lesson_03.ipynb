### 1. Задание
**1)** Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
Найдите их сумму. (на листочке)

**Решение:**  
x1 = (10, 10, 10)  
x2 = (0, 0, -10)  
x1 + x2 = (10, 10, 10) + (0, 0, -10) = (10 + 0, 10 + 0, 10 + (-10)) = (10, 10, 0)

**2)** Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
```

```python
def vector_len(x1, x2, x3):
    return np.sqrt(x1**2 + x2**2 + x3**2)
```

```python
vector_len(10, 10, 10)
```

```python
vector_len(4, 2, -4)
```

### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:  
1) окружности,  
2) эллипса,  
3) гиперболы.

```python
r = 5
x = [i / 100 for i in range(-1000, 1000, 1)]
y = [np.sqrt(r**2 - i**2) for i in x]
y1 = [-i for i in y]
plt.axis('equal')
plt.plot(x, y)
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y')
```

```python
a = 3
b = 4
x = [i / 100 for i in range(-1000, 1000, 1)]
y = [np.sqrt(b**2 * (a**2 - i**2)) / a**2 for i in x]
y1 = [-i for i in y]
plt.axis('equal')
plt.plot(x, y)
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y')
```

```python
a = 3
b = 4
x = [i / 100 for i in range(-1000, 1000, 1)]
y = [np.sqrt(b**2 * (i**2 - a**2)) / a**2 for i in x]
y1 = [-i for i in y]
plt.axis('equal')
plt.plot(x, y)
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y')
```

### 4. Задание (на листочке)  
**1)** Пусть задана плоскость:  
A * x + B * y + C * z + D = 0  
Напишите уравнение плоскости, параллельной данной и проходящей через начало
координат.

**Ответ:**  
Для параллельной плоскости, проходящей через начало координат D = 0:  
A * x + B * y + C * z = 0

**2)** Пусть задана плоскость: A1x + B1y + C1z + D1 = 0  
и прямая: (x-x1)/(x2-x1) = (y-y1)/(y2-y1) = (z-z1)/(z2-z1)  
Как узнать, принадлежит прямая плоскости или нет?

**Ответ**  
Прямая принадлежит плоскости, если две ее точки принадлежат этой плоскости. Т.е. подставив координаты двух точек этой прямой A(x1,y1,z1) и B(x2,y2,z2) в A1x + B1y + C1z + D1 = 0 мы должны получить верное равенство.

### 5. Задание (в программе)  
**1)** Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2 * X + 5 * Y
Z1 = 2 * X + 5 * Y + 155
ax.plot_wireframe(X, Y, Z, color='r')
ax.plot_wireframe(X, Y, Z1)
show()
```

**2)** Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
fig = figure()
ax = Axes3D(fig)
a = 5
b = 8
X = np.arange(-50, 50, 0.3)
Y = np.arange(-50, 50, 0.3)
X, Y = np.meshgrid(X, Y)
Z = X**2/a**2 + Y**2/b**2
ax.plot_wireframe(X, Y, Z)
show()
```

```python
fig = figure()
ax = Axes3D(fig)
r = 35
X = np.arange(-50, 50, 0.5)
Y = np.arange(-50, 50, 0.5)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(r**2 - X**2 - Y**2)
Z1 = -np.sqrt(r**2 - X**2 - Y**2)
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1)
show()
```

### 1. Задание (в программе)  
Нарисуйте график функции:  
y(x) = k·cos(x – a) + b  
для некоторых (2-3 различных) значений параметров k, a, b

```python
x = np.linspace(-5, 5, 1000)
y1 = 2 * cos(x - 30) + 2
y2 = 10 * cos(x - 60) - 2
y3 = 1/2 * cos(x - 45) + 25
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.grid(True)
plt.legend(labels=['y1', 'y2', 'y3'])
plt.xlabel('x')
plt.ylabel('y')
```
