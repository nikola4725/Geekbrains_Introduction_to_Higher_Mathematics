### 4. Задание (в программе)
**1)** Решите систему уравнений:  
y = x^2 – 1  
exp(x) + x*(1 – y) = 1

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-6, 6, 201)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def equations(p):
    x, y = p
    return(y - x**2 + 1, np.exp(x) + x - x * y - 1)
```

```python
x1, y1 = fsolve(equations, (-2, 2))
print(x1, y1)
```

```python
x2, y2 = fsolve(equations, (2, 4))
print(x2, y2)
```

```python
x3, y3 = fsolve(equations, (4, 4))
print(x3, y3)
```
